```python
import pandas as pd
from rdflib import Graph, Literal, RDF, URIRef, Namespace

# Load the CSV file
csv_file = 'Student_performance_data.csv'
df = pd.read_csv(csv_file)

# Define Namespaces
namespace = Namespace("http://www.mruniversity.edu/ontologies/curriculum-planner#")

# Create a Graph
g = Graph()

# Process the DataFrame and convert to RDF
for index, row in df.iterrows():
    student_uri = URIRef(namespace[f"Student_{index}"])
    g.add((student_uri, RDF.type, namespace.Student))
    g.add((student_uri, namespace.studentID, Literal(row['StudentID'])))
    g.add((student_uri, namespace.age, Literal(row['Age'])))
    g.add((student_uri, namespace.gender, Literal(row['Gender'])))
    g.add((student_uri, namespace.ethnicity, Literal(row['Ethnicity'])))
    g.add((student_uri, namespace.parentalEducation, Literal(row['ParentalEducation'])))
    g.add((student_uri, namespace.studyTimeWeekly, Literal(row['StudyTimeWeekly'])))
    g.add((student_uri, namespace.absences, Literal(row['Absences'])))
    g.add((student_uri, namespace.tutoring, Literal(row['Tutoring'])))
    g.add((student_uri, namespace.parentalSupport, Literal(row['ParentalSupport'])))
    g.add((student_uri, namespace.extracurricular, Literal(row['Extracurricular'])))
    g.add((student_uri, namespace.sports, Literal(row['Sports'])))
    g.add((student_uri, namespace.music, Literal(row['Music'])))
    g.add((student_uri, namespace.volunteering, Literal(row['Volunteering'])))
    g.add((student_uri, namespace.gpa, Literal(row['GPA'])))
    g.add((student_uri, namespace.gradeClass, Literal(row['GradeClass'])))

# Serialize the graph to a file
g.serialize('student_performance.rdf', format='turtle')
```

```python
from rdflib import Graph

# Load the RDF data
g = Graph()
g.parse('student_performance.rdf', format='turtle')

# Define a SPARQL query
query = """
PREFIX cp: <http://www.mruniversity.edu/ontologies/curriculum-planner#>
SELECT ?student ?id ?gpa
WHERE {
    ?student a cp:Student .
    ?student cp:studentID ?id .
    ?student cp:gpa ?gpa .
}
"""

# Execute the SPARQL query
results = g.query(query)

# Print the results
for row in results:
    print(f"Student URI: {row.student}, Student ID: {row.id}, GPA: {row.gpa}")
```
